### LAB 8-1: 오늘의 날짜와 현재 시간 출력하기

**문제 설명:**
오늘의 날짜와 현재 시간을 특정 형식에 맞게 출력합니다.
시간의 경우, 24시간제 형식을 확인하여 '오전' 또는 '오후' 정보를 함께 표시합니다.

```python
import datetime

now = datetime.datetime.now()

if now.hour < 12:
    ampm = "오전"
    hour = now.hour
else:
    ampm = "오후"
    hour = now.hour - 12 if now.hour > 12 else 12

print(f"오늘의 날짜 : {now.year}년 {now.month}월 {now.day}일")
print(f"현재시간 : {ampm} {hour}시 {now.minute}분 {now.second}초")
```

### LAB 8-2: 남은 날짜 계산하기 (1, 2, 3번 통합)

**문제 설명:**
객체 간의 차이 연산을 통해 특정 미래 시점(크리스마스, 새해, 생일)까지 남은 날짜와 시간을 계산합니다.

```python
import datetime

# 현재 시간 설정
today = datetime.datetime.now()
print(f"오늘은 {today.year}년 {today.month}월 {today.day}일입니다.")

# 1. 2027년 크리스마스까지 남은 시간
xmas_2027 = datetime.datetime(2027, 12, 25)
diff1 = xmas_2027 - today
print(f"2027년 크리스마스 까지는 {diff1.days}일 {diff1.seconds // 3600}시간 남았습니다.")

# 2. 2036년 새해까지 남은 시간
new_year_2036 = datetime.datetime(2036, 1, 1)
diff2 = new_year_2036 - today
print(f"2036년 새해 까지는 {diff2.days}일 {diff2.seconds // 3600}시간 남았습니다.")

# 3. 다가오는 생일(예: 2027년 5월 4일)까지 남은 시간
my_birthday = datetime.datetime(2027, 5, 4) 
diff3 = my_birthday - today
print(f"2027년 생일까지는 {diff3.days}일 {diff3.seconds // 3600}시간 남았습니다.")
```

### LAB 8-3: 날짜 계산하기

**문제 설명:**
1. 오늘로부터 1,000일 후의 날짜를 계산합니다.
2. 입력받은 기념일로부터 100일이 되는 날짜를 출력하는 `couple_day.py` 기능을 구현합니다.

```python
import datetime

today = datetime.datetime.now()
thousand_days = datetime.timedelta(days=1000)
future = today + thousand_days
print(f"오늘부터 1,000일 후의 날짜: {future.year}년 {future.month}월 {future.day}일")

print("-" * 30)

print("처음으로 사귄 연도와 월, 일을 띄어쓰기로 입력하시오 : ", end='')
year, month, day = map(int, input().split())

start_date = datetime.datetime(year, month, day)
hundred_days = datetime.timedelta(days=100)
anniversary = start_date + hundred_days

print(f"100일 기념일은 : {anniversary.year}년 {anniversary.month}월 {anniversary.day}일입니다.")
```

### LAB 8-4: math 모듈을 이용한 계산

**문제 설명:**
1. `math.pow()` 함수를 사용하여 4의 2승부터 10승까지 출력합니다.
2. `math.radians()` 함수를 사용하여 0도에서 180도까지 10도 단위로 라디안 값을 출력합니다.
3. `math.sin()` 함수를 사용하여 0도에서 180도까지 10도 간격으로 사인값을 출력합니다.

```python
import math

# 1. 4의 2승부터 10승까지 출력
print("1번 결과:")
for i in range(2, 11):
    result = math.pow(4, i)
    print(f"4**{i:2d} = {result}")

print("\n" + "-"*30 + "\n")

# 2. 일반각도(0~180도)를 라디안으로 변환하여 출력
print("2번 결과:")
for degree in range(0, 181, 10):
    radian = math.radians(degree)
    print(f"{degree:3d} degree = {radian:.3f} radian")

print("\n" + "-"*30 + "\n")

# 3. 0도에서 180도까지의 sin 값 출력 (10도 간격)
print("3번 결과:")
for degree in range(0, 181, 10):
    radian = math.radians(degree)
    sin_value = math.sin(radian)
    print(f"sin({degree:3d}) = {sin_value:.2f}")
```

### LAB 8-5: 랜덤 번호 생성기

**문제 설명:**
1. `random.randrange()`를 사용하여 0~100 사이의 5의 배수 중 3개를 무작위로 선택합니다.
2. `random.sample()`을 사용하여 1~10 사이의 정수 중 중복 없이 3개를 선택합니다.

```python
import random

# 1. 0에서 100 이하의 정수 중 5의 배수 3개 선택
five_multiples = [random.randrange(0, 101, 5) for _ in range(3)]
print(f"0에서 100 이하의 정수 중에서 5의 배수 {five_multiples}")

# 2. 1에서 10 사이의 정수 중 임의의 정수 3개를 sample()로 출력
sample_list = random.sample(range(1, 11), 3)
print(f"1에서 10 사이의 임의의 정수 : {sample_list}")
```

### LAB 8-6: 로또 번호 얻기

**문제 설명:**
1. 1~45 사이의 숫자 중 6개를 중복 없이 뽑는 로또 번호 생성을 5회 반복합니다.
2. 당첨 번호 6개와 보너스 번호 1개를 별도로 출력하는 프로그램을 작성합니다.

```python
import random

# 1. 매주 5개의 추천 로또 번호 생성
print("이번 주 추천 로또 번호")
for i in range(5):
    lotto = random.sample(range(1, 46), 6)
    lotto.sort()
    print(lotto)

print("\n" + "-"*30 + "\n")

# 2. 로또 당첨 번호 + 보너스 번호 출력
lotto_all = random.sample(range(1, 46), 7)
winning_numbers = sorted(lotto_all[:6])
bonus_number = [lotto_all[6]]

print(f"이번 주의 당첨 로또 번호 + 보너스")
print(f"{winning_numbers} + {bonus_number}")
```

### 연습문제 8.1: 리스트 합계와 평균 구하기

**문제 설명:**
사용자로부터 5개의 정수를 입력받아 리스트에 저장한 후, 리스트 내 모든 요소의 합계와 평균을 계산하여 출력합니다.

```python
# 정수 5개를 입력받아 리스트 생성
numbers = list(map(int, input("정수 5개를 띄어쓰기로 구분하여 입력하세요: ").split()))

# 합계와 평균 계산
total_sum = sum(numbers)
average = total_sum / len(numbers)

# 결과 출력
print(f"입력된 리스트: {numbers}")
print(f"합계: {total_sum}")
print(f"평균: {average}")
```

### 연습문제 8.2: 최댓값과 최솟값 찾기

**문제 설명:**
임의의 정수들이 담긴 리스트에서 가장 큰 값(최댓값)과 가장 작은 값(최솟값)을 찾아 출력합니다.

```python
import random

# 1~100 사이의 난수 10개로 리스트 생성
data = random.sample(range(1, 101), 10)

# 최댓값과 최솟값 계산
max_val = max(data)
min_val = min(data)

# 결과 출력
print(f"생성된 리스트: {data}")
print(f"최댓값: {max_val}")
print(f"최솟값: {min_val}")
```

### 연습문제 8.3: 중복 원소 제거

**문제 설명:**
중복된 원소가 포함된 리스트에서 중복을 제거하고 유일한 원소들만 남긴 리스트를 생성합니다.

```python
# 중복이 포함된 리스트
original_list = [10, 20, 10, 30, 40, 30, 50, 20, 60]

# set 자료형을 이용하여 중복 제거 후 다시 리스트로 변환
unique_list = list(set(original_list))
unique_list.sort() # 보기 좋게 정렬

# 결과 출력
print(f"원본 리스트: {original_list}")
print(f"중복 제거 후 리스트: {unique_list}")
```

### 연습문제 8.4: 2차원 리스트 행의 합

**문제 설명:**
3x3 크기의 2차원 리스트를 생성하고, 각 행의 합계를 구하여 출력합니다.

```python
# 3x3 2차원 리스트 생성
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print("--- 행렬 출력 ---")
for row in matrix:
    print(row)

print("\n--- 각 행의 합계 ---")
for i, row in enumerate(matrix):
    row_sum = sum(row)
    print(f"{i+1}행의 합계: {row_sum}")
```
